```python
#Eigen Values & Eigen Vectors

import numpy as np

A = np.array([[1, 2],
              [2, 1]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)    #also can use .eigh() which is specialized for hermitian matrices and is more stable, faster and gurantees purely real eigen values

print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors Matrix:\n", eigenvectors)

#Proof A(v) = lambda(v)
L = eigenvalues[0]
v = eigenvectors[:,0]   #eigenvectors are stacked as columns and are normalized automatically before returning

Av = A @ v
Lv = L * v

print("\nA @ v:", Av)
print("lambda * v:", Lv)
print("Match?:", np.allclose(Av, Lv))

#NOTE: for certain arrays, eigenvalues and eigenvectors may contain complex numbers so numoy automatically shifts the output array dtype to complex128
```

```python
#SVD: Singular Value Decomposition: Unlike Eigen values which can only decompose square matrix, SVD is used to decompose any rectangular matrix into three parts.

# A non-square 3x2 matrix
A = np.array([[1, 2],
              [3, 4],
              [5, 6]], dtype=float)

# Compute SVD
U, S, Vh = np.linalg.svd(A)

print("U Matrix Shape: ", U.shape)
print("Singular Values (S):", S)
print("Vh Matrix Shape:", Vh.shape)
```
