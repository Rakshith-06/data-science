```python
#Modern Random Engine:

import numpy as np

#initializing the modern random number generator engine
rng = np.random.default_rng(seed=42) #seed parameter (default = None) allows us to set a value from which it starts generating numbers.
                                     #It ensures that the value generated do not change until and unless:
                                        # 1)Your numpy version is changed
                                        # 2)You change the order of your code (by adding or removing lines from it)
                                        # 3)You run your program by multithreading
                                     #Two different generators with the same seed generate the same values if the above conditions are satisfied 

# Generate 3 uniform random floats between 0.0 (inclusive) and 1.0 (exclusive)
random_floats = rng.random(size=3)
print("Uniform Floats: ", random_floats)

# Generate a 2x3 matrix of random integers between 1 and 100
random_ints = rng.integers(low=1, high=101, size=(2, 3))
print("\nRandom Integers Matrix:\n", random_ints)

#Pre-defining mean(centre) and standard deviation(spread)
normal_samples = rng.normal(loc=0.0,scale=1.0,size=(3,5))
norm_acc_samples = rng.normal(loc=0.0,scale=1.0,size=1000)     #this result is more accurate because it has 1000 elements in it while the array before has only 15
print("Mean =",np.mean(normal_samples))
print("More Accurate Mean =",np.mean(norm_acc_samples))
print("Std Deviation =",np.std(normal_samples))
print("More Accurate Std Deviation =",np.std(norm_acc_samples))
```

```python
#Some more functions of Modern random generator np.random.default_rng()

#choice(): choosing among a group of choices with custom probability.
outcomes = ['Heads', 'Tails']
print(rng.choice(outcomes, size=5, p=[0.7, 0.3]))  #returns an array of 5 choices with heads having 70% probability to occur


#shuffle(): randomly shuffles the array in-place
arr = np.array([10, 20, 30, 40, 50])
rng.shuffle(arr)
print(arr)

#permutation(): randomly shuffles the array and returns a copy of the same.
arr = np.array([10, 20, 30, 40, 50])
random_arr = rng.permutation(arr)
print(random_arr)
```
