```python
#ndim: The number of dimensions (axis)
#shpae: Tuple showing number of elements along each axis
#size: Total number of elements in it

import numpy as np

# Create a 2x4 matrix of floats
matrix = np.array([[1.5, 2.7, 3.9, 4.1], 
                   [5.0, 6.2, 7.4, 8.6]], dtype=np.float64)

print("=== 1. Structural Attributes ===")
print("Dimensions (ndim):", matrix.ndim)
print("Shape (shape):     ", matrix.shape)
print("Total Items (size):", matrix.size)

print("\n=== 2. Memory Footprint ===") #Memory footprint refers to the exact amount of computer RAM that the program is using while it runs.
#Memory Footprint: nbytes = size of array * size of individual element in it
print("Data Type (dtype): ", matrix.dtype)
print("Bytes per item:    ", matrix.itemsize)
print("Total array bytes: ", matrix.nbytes)

# Verifying the interview equation
#assert is a keyword used to check of a condition is True or not. if false it crashes your program and raises an AssertionError
assert matrix.nbytes == matrix.size * matrix.itemsize
print("-> Mathematical formula verified successfully!")

print("\n=== 3. Type Conversion (Casting) ===")
# Cast floats to 32-bit integers (truncates decimals!)
# Note: astype() creates a new copy of the array and does not changes the data type in place.
int_matrix = matrix.astype(np.int32)

print("New Data Type:     ", int_matrix.dtype)
print("New Item Size:     ", int_matrix.itemsize, "bytes")
print("New Total Bytes:   ", int_matrix.nbytes, "bytes")
print("Truncated Values:\n", int_matrix)
```
