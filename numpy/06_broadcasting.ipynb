```python
#broadcasting: It allows arithmetic operations to be performed between arrays of different shapes 
#without creating unnecessary copies of the data in memory. For two arrays to be compatible for broadcasting
    #they are completely of the same size   
    #one of the dimensions is 1

import numpy as np

arr = np.array([1, 2, 3])
result = arr * 2    #numpy treats 2 as an array of shape (1,) -> [2] and stretches ot to array of shape (3,) -> [2,2,2]
print(result)

#operation among 1D and 2D arrays
matrix = np.array([[10, 20, 30],[40, 50, 60]])  # Shape: (2, 3)
row_vector = np.array([1, 2, 3])   # Shape: (3,)

result = matrix + row_vector
print("\n",result)

#broadcasting two vectors together
col_vec = np.array([[1], [2], [3]]) # Shape: (3, 1)                     => stretched to [[1,1],[2,2],[3,3]]
row_vec = np.array([10, 20])        # Shape: (2,) -> padded to (1, 2)   => stretched to [[10,20],[10,20],[10,20]]

result = col_vec + row_vec
print("\n",result)
```
