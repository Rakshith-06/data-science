```python
import numpy as np

# Define our related arrays
features = np.array([[1.5, 2.3], [4.1, 5.6]])
labels = np.array([0, 1])

# Save multiple arrays using key-value pair assignments
np.savez('dataset_archive.npz', X=features, y=labels)

# Load the archive back into memory
archive = np.load('dataset_archive.npz')

# View the stored key names (works like a Python dictionary)
print("Keys in archive:", archive.files)

# Extract specific arrays by their key name
print("Features (X):\n", archive['X'])
print("Labels (y):  ", archive['y'])

# Always close the file handle when done!
archive.close()
```
