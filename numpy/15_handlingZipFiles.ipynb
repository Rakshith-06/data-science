```python
import numpy as np

#np.savez(): saves multiple logically-related arrays in a binary file (.npz) using key-value pair assignments. 
# A .npz file is used because it acts as a zipped folder (archive) that bundles multiple .npy files together, 
# whereas a single .npy file is structurally limited to storing only one array.
features = np.array([[1.5,2.7],[3.4,4.9]])
labels = np.array([0,1])

np.savez('dataset.npz',X=features,y=labels) #field name can be anything. If not passed then Python would name it itself any random name like arr_0 etc

archive = np.load('dataset.npz')    #this performs lazy loading: doesn't loads the entire array until specified

print("The key values in the archive are:",archive.files)

print("Features:",archive['X'])
print("Labels:",archive['y'])

#always remember to close the archive. In case of np.save() the file closes the file automatically until you pass an already opened file.
archive.close()
```

```python
#np.savez_compressed(): whenever some data is being repeated in our file several times and we want to losslessly compress it we use this.
sparse_matrix = np.zeros((1000,1000))
sparse_matrix[50,50] = 99.9 #keeping a value non-zero

np.savez_compressed('compressed_archive.npz',matrix=sparse_matrix)

loaded_matrix = np.load('compressed_archive.npz')
print("The loaded matrix has a field name of:",loaded_matrix.files)

#NOTE: You can actually rename a .npz extension to .zip and unpack it using standard OS file extractors to reveal the raw .npy files inside!
```

```python
# Question 1: Create two 1D arrays: a = np.arange(5) and b = np.arange(10). 
# Save them to an archive named my_data.npz using custom keys first and second.
a = np.arange(5)
b = np.arange(10)

np.savez('pracOfZipFiles.npz',first=a,second=b)

with np.load('pracOfZipFiles.npz') as f:
    a1 = f['first']
    b1 = f['second']
    print("The two matrices extracted from the file are:\na:",a1,"\nb:",b1)

# Question 2: Write a snippet using a with statement to load my_data.npz and print the sum of array second.
with np.load('pracOfZipFiles.npz') as f:
    arr = f['second']
    Sum = np.sum(arr)

print("\nThe required sum is:",Sum)
```
