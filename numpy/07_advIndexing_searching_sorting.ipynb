```python
#np.where(): helps us to scan an array for a condition and returns the values if condition is True
import numpy as np

arr = np.array([10,5,20,3,15])
result = np.where(arr > 10, arr, -1) #np.where(<condition>, <value to return if cond True>, <value to return if cond False>)
print(result)

#only giving condition 
arr = np.array([10, 5, 20, 3, 20])
indices = np.where(arr == 20)   #returns a tuple of arrays containing the index positions where the condition evaluates to True
print("\n",indices)

#array passed itself
arr = np.array([10,0,2,-37,0])
result = np.where(arr)  #returns indices of all truthly(non-zero) values in the array
print("\nResult:",result)
```

```python
# Question 1: Given scores = np.array([45, 88, 72, 30, 95]), use np.where to create an array containing "Pass" for scores >= 50 and "Fail" otherwise.
scores = np.array([45,88,72,30,95]) 
result = np.where(scores>=50,"Pass","Fail")
print("Result:",result)

# Question 2: Find the coordinate indices of all negative numbers in a 2D matrix.
arr = np.array([[15, -3, 8],[ -7, 0, 12],[ 4, -19, -5]])
coord = np.where(arr<0)
print("\nThe indices of all the negative numbers is:",coord)  #the tuple has two tuples 1)row indices 2)column indices
```

```python
#sorting: sorting can be done in two ways:
    #1> np.sort(<arr>) -> sorts the array in-place 

import numpy as np

arr = np.array([10,2,34,28,17])
sort_arr = np.sort(arr)
print("Sorted array:",sort_arr)

    #2> np.argsort(<arr>) -> returns a new array containing the order of indices of the sorted arrangement of the array

arr = np.array([10,2,37,27,19])
sorted_ind = np.argsort(arr)
print("\nThe indices in sorted form are:",sorted_ind)
print("The sorted array is:",arr[sorted_ind])
print("while the original array still remains:",arr)

matrix = np.array([[15,2],[32,14]])
sorted_matrix = np.sort(matrix)
arg_sort_matrix = np.argsort(matrix)
print("\nSorting 2D array using sort():\n",sorted_matrix)
print("\nSorting 2D array using argsort():\n",arg_sort_matrix)

#IMP NOTE: np.sort(arr) v\s arr.sort()

#np.sort(arr): it is a function which returns a new copy of the sorted array
arr = np.array([20,3,60,27,86]) 
sort_arr = np.sort(arr)
print("\nThe result of np.sort() is the sorted array:",sort_arr)
print("Original array after sorting:",arr)

#arr.sort(): it is a method which sorts the array in-place and returns None
arr = np.array([40,37,20,57,2,13])
result = arr.sort()
print("\nThe result of arr.sort():",result,"and sorted array:",arr)
print("Original array:",arr)

#sorting in descending order

#using np.sort()
arr = np.array([40,37,20,57,2,13])
des_arr = np.sort(arr)[::-1]    #reverse the sorted array
print("\nDescending order of array:",des_arr)

#using np.argsort()
arr = np.array([40,37,20,57,2,13])
des_arr = np.argsort(-arr)  #since it returns only indices and the more magnitude of -ve numbers the least is considered their value 
print("Descending order of array:",arr[des_arr])
```

```python
# Question 1: Given arr = np.array([12, 4, 8, 15, 1]), use np.argsort() to find the index of the second smallest element.
arr = np.array([12, 4, 8, 15, 1])
sort_ind = np.argsort(arr)
print("The index of the second smallest number is:",sort_ind[1])

# Question 2: Create a 2D matrix of shape (3, 3) with random integers and sort it along its columns (axis=0).
arr = np.array([[10,27,1],[2,40,37],[36,5,19]])
sort_arr = np.sort(arr,axis=0)
print("Original array:\n",arr)
print("The sorted matrix in column form is:\n",sort_arr)
```

```python
#np.searchsorted(): returns an array of indices where the new elements have to be inserted in order to maintane their sorted order.

arr = np.array([1,15,20,37,69,102])
index = np.searchsorted(arr,[23,34,110])
print("The indices at which [23, 34, 110] have to be inserted are:",index)

#23 must come at 3rd index -> between 20 and 37
#34 must come at 3rd index -> between 20 and 37
#110 must come at 6th index -> at the end after 102

#IMP NOTE: if your source array isn't already sorted, it will return incorrect insertion indices without throwing an error.
        #  it uses binary search to find location taking O(log(n)) time

#np.unique(): returns a new array of all unique elements in a sorted manner
arr = np.array([38,2,49,16,2,43,16,92,2,16])
uniq = np.unique(arr)
print("\nOriginal array:",arr)
print("Unique array:",uniq)

#works also with strings
data = np.array(['cat', 'dog', 'cat', 'bird', 'dog', 'dog'])
unique_vals, counts = np.unique(data, return_counts=True)   #it returns two things in the form of a tuple: 1) the sorted set of elements 
print("Unique Elements:", unique_vals)                                                                  #  2) the number of times each value is repeated 
print("Counts:", counts)                                    #there also is a parameter named return_index which when set to True returns the indices of the first appearance of an element in an array

#unique matrices
arr = np.array([[1,2,3,2,1],[4,6,5,5,6],[7,9,8,7,9]])
uniq_mat = np.unique(arr)
print("\n",uniq_mat)     #it flattens the array if no parameter passed 
uniq_row = np.unique(arr,axis=1)
print("\n",uniq_row)
```
