```python
#Boolean Masking: highly optimized extraction technique which eliminates the use of slow python for loops and if conditions

import numpy as np

# A collection of daily temperatures
temperatures = np.array([72, 68, 85, 92, 64, 78, 89])

# 1. Create the boolean mask array (evaluates element-wise)
mask = temperatures > 80
print("The Mask Array:", mask)

# 2. Apply the mask to extract matching elements
hot_days = temperatures[mask]
print("Filtered Output:", hot_days)


#Multi-conditioning with bitwise operators
ages = np.array([12, 25, 17, 40, 55, 19])

# Extract values between 18 AND 50
# Crucial: You MUST wrap each separate condition inside parentheses!
young_adults_mask = (ages >= 18) & (ages <= 50)
print("\nFiltered Ages:", ages[young_adults_mask])
```

```python
# Question 1: Given arr = np.array([-2, 5, 0, -8, 3]), write a snippet to replace all negative numbers with exactly 0 using a boolean mask.
arr = np.array([-2,5,0,-8,3])
negat_arr = arr < 0
arr[negat_arr] = 0
print("New replaced values are:",arr)

# Question 2: Create a 2D matrix of shape (3, 3) containing numbers 1 to 9. Extract all elements that are perfectly divisible by 3 (% 3 == 0).
arr = np.arange(1,10).reshape(3,3)
Arr = arr % 3 == 0
print("All elements perfectly divisble by 3 are:",arr[Arr])
```
