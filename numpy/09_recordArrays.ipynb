```python
#record arrays: it allows you to access columns/fields using attribute dot-notation (array.field) instead of standard string keys (array['field']).
#it is a subclass of Python Numpy

import numpy as np

# Define our structured data type layout
student_dtype = [('name', 'U10'), ('grade', 'i4')]

# Create a standard structured array first
structured_arr = np.array([('Sam', 85), ('Alex', 92)], dtype=student_dtype)

# Convert it directly to a Record Array
records = structured_arr.view(np.recarray)

# Access fields seamlessly using dot-notation! Also standard key indexing can also be used: records[grade]
print("All Grades:", records.grade)
print("First Name:", records.name[0])

# although record arrays look a lot more cleaner than structured arrays but they are slower than them since while calling:
# records[grade] ->  NumPy immediately triggers a direct, compiled C-level dictionary lookup
# records.grade -> Python must invoke the __getattributes__ method which intercepts the call, searches the object's instance dictionary,
# evaluates whether the field name matches or not and then routes accordingly

# If your custom field name matches a built-in NumPy array method name (like shape, size, or mean), 
# using the dot-notation will call the built-in array method instead of returning your column data
```

```python
# Question 1: Create a record array with a field called price containing a list of float values. Access the column using dot-notation and compute its average using .mean().
price = [("Float1","f8"),("Float2","f8"),("float","f8")]
data = [(10.0,100.0,11.0),(20.9,223.2,40.5),(310.1,35.3,30.1)]
arr = np.array(data,dtype=price)
arr = arr.view(np.recarray)
print("The mean of:",arr.Float1,"is:",arr.Float1.mean())

# Question 2: Convert a standard structured array with fields x and y into a record array using np.recarray.
structured_arr = np.array([(1, 10.5), (2, 20.5)], dtype=[('x', 'i4'), ('y', 'f4')])
rec_arr = np.rec.array(structured_arr)
print(rec_arr.x) 
print(rec_arr.y)  
```
