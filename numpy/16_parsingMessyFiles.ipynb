```python
# np.loadtxt(): is a fast, lightweight parser that works great for clean, well formated numerical text files.
# But once we face a messy file(file with disorganized, poorly formated or inconsistent data) we can't use this method anymore.

# np.genfromtxt(): hence we have this method for parsing messy files

import numpy as np

# inconsistent data in file = [[1.5,2.1,3.7],
#                              [3.3,4.2,___],
#                              [5.3,6.0,7.9]]

data_loaded =  np.genfromtxt('parsingMessyFiles.csv',delimiter=',',filling_values=-1)   # default value of filling_values is NaN (Not a Number)
                                                                                        # any operation performed with NaN always results in NaN

#if we had used loadtxt() then it would have returned a ValueError
print("Parsed data:\n",data_loaded)

#IMP NOTE: np.loadtxt() is much more faster than np.genfromtxt() for larger datasets since it assumes that the file being passed is already clean, well formated. 
```
