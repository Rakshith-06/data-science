```python
#Permuting matrix: for matrices of 2 dim, to change the order of elements we use np.transpose() function. 
#                  But for matrices with dimesnions more than 3, we use permute(pass a tuple with new dimensions in it)

import numpy as np

# A tensor is a multi-dimensional grid of numbers
tensor = np.ones((2, 3, 4))

permuted = np.transpose(tensor, axes=(1, 2, 0)) # Permute axes: Axis 1 becomes Axis 0, Axis 2 becomes Axis 1, Axis 0 becomes Axis 2
#by default it transposes the matrice's shape from (a,b,c,....x,y,z) to (z,y,x....c,b,a)
print("Original tensor:\n",tensor)
print("Original tensor's shape:",tensor.shape)
print("\nPermuted shape:\n",permuted)
print("Permuted Shape:", permuted.shape)

#Calculating the Hermitian Transpose (Transpose Conjugate[Tranjugate])
matrix = np.array([[1+2j,3-4j,5],[6j,7,8+1j]])
tranj_matrix = matrix.conj().T
print("\nThe conjugate transpose of the given matrix is:",tranj_matrix)
```

```python
#determinant of matrix
A = np.array([[4,7],[2,6]])
det = np.linalg.det(A)
print("The determinant of the given matrix is:",det)

#determinant of multi-dimensional arrays
matrix = np.array(
    [[[1, 2, 3],
      [0, 4, 5],
      [1, 0, 6]],
      
      [[2, 5, 1],
       [3, 0, 2],
       [4, 1, 3]]]
)
#it will simply treat it as a batch of 2 3x3 matrices
det = np.linalg.det(matrix)
print("\nThe determinant of the given matrix of shape:",matrix.shape,"is:",det)

#inverse of a matrix
inv = np.linalg.inv(A)
print("\nThe inverse of the given matrix:\n",inv)

#if the matrix is Singular (i.e det(A) = 0) then np.linalg.inv(A) returns LinAlgError: Singular Matrix

#IMP NOTE: Solving sets of linear equations Ax = b
#Generallyt these questions are solved by first finding out the value of inv(A) and then multiplying it with b, we get
# x = b.inv(A)
#but using x = np.linalg.inv(A)@b is a very expensive operation to perform so we instead use np.linalg.solve(A,b)

#let equations be:
    # x+2y+3z=14
    # 2x+5y+2z=18
    # x+3y+z=11

A = np.array([[1,2,3],[2,5,2],[1,3,1]])
b = np.array([14,18,11])

x,y,z = np.linalg.solve(A,b)
print("\nThe values of x:",x,", y:",y,"and z:",z)
```

```python
#Eigen Values & Eigen Vectors

A = np.array([[1, 2],
              [2, 1]])

# Compute eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)    #also can use .eigh() which is specialized for hermitian matrices and is more stable, faster and gurantees purely real eigen values

print("Eigenvalues:\n", eigenvalues)
print("\nEigenvectors Matrix:\n", eigenvectors)

#Proof A(v) = lambda(v)
L = eigenvalues[0]
v = eigenvectors[:,0]   #eigenvectors are stacked as columns 

Av = A @ v
Lv = L * v

print("\nA @ v:", Av)
print("lambda * v:", Lv)
print("Match?:", np.allclose(Av, Lv))

#NOTE: for certain arrays, eigenvalues and eigenvectors may contain complex numbers so numoy automatically shifts the output array dtype to complex128
```
