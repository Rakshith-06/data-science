```python
#Structured Arrays: allow us to define custom data types composed of multiple named fields, each with its own data type

import numpy as np

#1. Defining a custom data type: employee_dtype
employee_dtype = [('name','U10'),('age','i4'),('salary','f4')]
#U10: unicode string of max length 10
#i4: 4-byte integer
#f4: 4-byte float

#2. Creating an array of the custom data type
data = [("Alice",32,75000.0),("Bob",28,62000.0),("Charlie",45,90000.0)]

emp = np.array(data, dtype=employee_dtype)
print(emp)
print(type(emp))

#Accessing values in different fields
print("\nName of all employees:",emp['name'],"and their salary is:",emp['salary'])  #using field name returns a fast memory view and not creates a new array
print("The first entry of the array is:",emp[0])

#Conditioning in structured arrays
highly_paid = emp[emp['salary']>70000]
print("\nHighly paid employees are:",highly_paid['name'])

#Both emp[0]['name'] and emp['name'][0] would return the same thing ("Alice") but 
#emp[0]['name'] format -> Best when focusing on a specific individual's data.
#emp['name'][0] format -> Best when extracting data for bulk or vectorized processing.
```
